* This example uses prebuilt linear regression function
* And the dataset provided is also prebuilt without any hassle
* it is just a fancy way of finding out the best straight line which will describe the relation of the variable

```python
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Loading a prebuild dataset from sklearn
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
type(diabetes_X)
print(diabetes_X.shape)
print(diabetes_y.shape)
print(diabetes_X)
```

```python
diabetes_X = diabetes_X[:,np.newaxis,2]
print(diabetes_X)
```

```python
#split the data into training / testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
print(diabetes_X)
```

```python
#split the data into training / testing sets
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
```

```python
regr = linear_model.LinearRegression()
regr.fit(diabetes_X_train,diabetes_y_train)

diabetes_y_pred = regr.predict(diabetes_X_test)

# the coefficients
print("Coefficients:  {}\n ".format(regr.coef_))

# The mean squared error
print("Mean Squared error : {}".format(mean_squared_error(diabetes_y_test,diabetes_y_pred)))
# The coefficient of determination 1 is perfect prediction
print("Coefficient of determination is {}".format(r2_score(diabetes_y_test,diabetes_y_pred)))
```

```python
plt.scatter(diabetes_X_test,diabetes_y_test,color="black")
plt.plot(diabetes_X_test,diabetes_y_pred,color="blue",linewidth=3)

plt.xticks(())
plt.yticks(())
plt.show()
```
